```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
import requests
import selenium.webdriver.chrome.webdriver
import time, os
from PIL import Image
import numpy as np

URL = "https://mofanpy.com/static/scraping/basic-structure.html"
URL = 'https://www.cdc.gov.tw/Bulletin/List/MmgtpeidAR5Ooai4-fgHzQ?page=1&fbclid=IwAR1k7Axr3Nt6zPNHUdu1VYOo1OV1xOv4l5EkfVfbSXr9VZakbjUOxjGbICY'
```

```python
# 設定想爬的頁數: ex n_i = 1/n_f =9 表示我要爬第1~9頁
n_i = 20
n_f =30
```

```python
def find_date():
    text_right_info = soup.find_all('div', class_="text-right")
    date_info  =text_right_info[0]
    date_info_cut = '{}'.format(date_info)
    if '發佈日期：' in date_info_cut:
        date = date_info_cut[date_info_cut.index('發佈日期：')+5:-6]
        return date
    else:
        return None
    
def find_article_url_in_outline_url(URL):
    html = urlopen(URL).read().decode('utf-8')
    soup = BeautifulSoup(html, features='lxml')
    city_tags = soup.find_all('div', class_="content-boxes-v3")
    cities_links = [temp_url + tag.find('a').get('href','') if tag.find('a') else '' for tag in city_tags]
    
    return cities_links
```

```python
All_News_url=[]

if n_i == 1:
    temp_url = 'https://www.cdc.gov.tw/Bulletin/List/MmgtpeidAR5Ooai4-fgHzQ'
    cities_links = find_article_url_in_outline_url(temp_url)
    for j in range(len(cities_links)):
        temp_content = cities_links[j][:32]+'Detail/'+cities_links[j][76:]
        All_News_url.append(temp_content)
        print("1 :",temp_content)    
    
    for page in range(n_i+1,n_f+1):
        temp_url = 'https://www.cdc.gov.tw/Bulletin/List/MmgtpeidAR5Ooai4-fgHzQ?page={}'.format(page)
        cities_links = find_article_url_in_outline_url(temp_url)
    
        for j in range(len(cities_links)):
            temp_content = cities_links[j][:32]+cities_links[j][76:]
            All_News_url.append(temp_content)
            print(page,":",temp_content)
else:
    for page in range(n_i,n_f+1):
        temp_url = 'https://www.cdc.gov.tw/Bulletin/List/MmgtpeidAR5Ooai4-fgHzQ?page={}'.format(page)
        cities_links = find_article_url_in_outline_url(temp_url)
    
        for j in range(len(cities_links)):
            temp_content = cities_links[j][:32]+cities_links[j][76:]
            All_News_url.append(temp_content)
            print(page,":",temp_content)
```

```python
for i in range(len(All_News_url)):
    temp_url = All_News_url[i]
    print(temp_url)
    html = urlopen(temp_url).read().decode('utf-8')
    soup = BeautifulSoup(html, features='lxml')
    date  = find_date()
    
    temp_each_article = ''
    ###########################################################################
    print(soup, file=open('file.txt', 'w', encoding='UTF-8')) # 這條不可以刪!!!!
    ###########################################################################
    with open('file.txt','r',encoding='utf-8') as fh:
        tmp = fh.read()
        onelinelist = tmp.split('\n')
        kk = 0
        n = 0
        for k in range(len(onelinelist)):
            sub_element = onelinelist[k]
            len_sub_element = len(sub_element)
            for s in range(len_sub_element):
                if (sub_element[s] =='b') and (sub_element[s+1] =='r'):
                    if n == 0:
                        kk = k
                    if kk != k:
                        temp_each_article = temp_each_article + '{}'.format(sub_element[:s-1])
                    n += 1
                    
    temp_each_article =  temp_each_article[37:]
    writepath = 'articles/No{}_date_{}.txt'.format(i+1,date)
    print("Date:", date)
    print(temp_each_article)
    print("\n")
    f = open(writepath, 'w', encoding='UTF-8')
    f.writelines('[Article url]: \n')
    f.writelines(temp_url)
    f.writelines('\n\n')
    f.writelines('[Content]\n')
    f.writelines(temp_each_article)
    f.close()
```
